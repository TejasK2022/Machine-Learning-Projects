```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('DATASETS/student_clustering.csv')
```

```python
print('Shape of the data is :- ',df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['cgpa'],df['iq'])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.show()
```

```python
# As we clearly see that there is 4 clusters but lets test it on elbow method
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,10):
    km=KMeans(n_clusters=i) # Here we define the number of clusters = i
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)  # This is elbow method lets see it also show it becomes constant at 4
plt.show()
```

```python
X=df.iloc[:,:].values # Just converting the dataframe into array for ease calculation
km=KMeans(n_clusters=4) # Lets choose number of clusters=4
y_means=km.fit_predict(X) # It will assign the cluster to each point
y_means
```

```python
# Now i have to plot the graph 
plt.scatter(X[y_means==0,0],X[y_means==0,1],color='blue')
plt.scatter(X[y_means==1,0],X[y_means==1,1],color='red')
plt.scatter(X[y_means==2,0],X[y_means==2,1],color='green')
plt.scatter(X[y_means==3,0],X[y_means==3,1],color='yellow')
```
