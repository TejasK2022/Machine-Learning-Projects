```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=pd.read_csv('E:\\MACHINE LEARNING TEJAS\\spam_email.csv', encoding='latin-1')
```

```python
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True) # dropping column having no use in this example
```

```python
df.isna().sum()
```

```python
X=df.iloc[:,0]
y=df.iloc[:,-1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(X)
```

```python
plt.scatter(df['Text'].head(50),df['Spam/Ham'].head(50))
plt.xlabel('Messages')
plt.ylabel('Spam or Not')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,r2_score,confusion_matrix
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
